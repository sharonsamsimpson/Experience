# Data Scrapping from MEDLINE PLUS website

__Website Link =>__ https://medlineplus.gov/healthtopics.html

```python
import re
import json
from string import punctuation
import urllib.request as ur
from bs4 import BeautifulSoup as bs
```

```python
def clean_data(text):
    text = text.lower()
    nopunc = [char for char in text if char not in punctuation]
    nopunc = ''.join(nopunc)
    result = [word for word in nopunc.split()]
    result = ' '.join(result)
    return result
```

```python
def disease_scrapper():
    
    dictionary = {}
    dictionary['Disease_Data'] = []
      
    for alphabets in range(ord('a'), ord('b') + 1):
    
        url_alphabets = "https://medlineplus.gov/healthtopics_" + chr(alphabets) + ".html"
        print('#########################     Getting diseases starting with', chr(alphabets) ,'   ############################')
        html_alphabets = ur.urlopen(url_alphabets).read()
        soup_alphabets = bs(html_alphabets, 'html.parser')
        article_list = soup_alphabets.find('article')

        for ul_tag in article_list('ul'):
            for li_tag in ul_tag('li'):
                for a_href in li_tag('a'):
                    
                    url_diseases = a_href['href']
                    disease_name = clean_data(a_href.text)
                    html_diseases = ur.urlopen(url_diseases).read()
                    soup_diseases = bs(html_diseases, 'html.parser')
                    text_diseases = soup_diseases.find('div', attrs={"id":"topic-summary", "class":"syndicate"})
                                
                    bs_text_diseases = bs(str(text_diseases), "lxml").text
                    clean_text_diseases = []
                    clean_text_diseases.append(clean_data(bs_text_diseases))

                    disease_object = {disease_name : clean_text_diseases}
                    dictionary['Disease_Data'].append(disease_object)
                    
                    with open('Disease_Data.json', 'w') as json_file:
                        json.dump(dictionary, json_file, indent=4)
                        print(disease_name)
```

```python
try:
    disease_scrapper()
    
except Exception as er:
    print(er)
```
