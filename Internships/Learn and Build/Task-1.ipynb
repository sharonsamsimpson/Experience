```python
# Create Phone Directory
print("                    Welcome to Phone Directory")
print('---------------------------------------------------------------------')

# Create a phone directory with menu
def menu():
    menu = input("Select \n 'a' to add \n 'u' to update \n 'd' to delete \n 's' to select \n 'sa' to select all \n ")
    
    if menu == 'a':
        add_contact()
    elif menu == 'u':
        update_contact()
    elif menu == 'd':
        delete_contact()
    elif menu == 's':
        select_contact()
    elif menu == 'sa':
        select_all_contact()
    else:
        print('Invalid response')
 
 
def main():
    while True:
        menu()
        Quit = input("Type 'n' to quit or 'y' to stay: ")
        if Quit == 'y':
            pass
        elif Quit == 'n':
            break
        else:
            print('Invalid response')


# create contact
contact = {'Allen':'6726319932', 'Alice':'7849182815', 'Sherin':'7438999742', 'Sam':'4892092910', 'Sweety':'8939902881'}
    
# add new contact
def add_contact():
    print("Enter the details to ADD a new contact")
    name = str(input("Enter Name: "))
    number = str(input("Enter Number: "))
    add = False
    
    for i in contact:
        if i == name:
            print('\n')
            print('Name already there')
        else:
            add = True
    if add:
        contact[name] = number
        print('\n')
        print('New Contact Added')
    print('---------------------------------------------------------------------')

# update contact
def update_contact():
    print("Enter the details to UPDATE a contact")
    name = str(input("Enter Name: "))
    number = str(input("Enter Number: "))
    
    if name in contact.keys():
        print('Contact exists', end =" ")
        contact.update({name : number})
        print("\n")
        print('Contact Updated')       
    else:
        print('\n')
        print("Contact doesn't exists")  
    print('---------------------------------------------------------------------')
        
# delete contact    
def delete_contact():
    print("Enter the details to DELETE a contact")
    name = str(input("Enter Name: "))
    if name in contact.keys():
        contact.pop(name)
        print('\n')
        print('Contact Deleted') 
    else:
        print('\n')
        print("Contact doesn't exists")
    print('---------------------------------------------------------------------')

# show a give name contact
def select_contact():
    print("Enter the details to SELECT a contact")
    name = str(input("Enter Name: "))
    if name in contact.keys():
        for key, value in contact.items():
            if key == name:
                print(f'Name: {key} ----- Number: {value}')
        print('\n')
        print('Contact Selected') 
    else:
        print('\n')
        print("Contact doesn't exists")
    print('---------------------------------------------------------------------')
    
# show all contact
def select_all_contact():
    print("Enter the details to SELECT all the contacts")
    print('\n')
    print('All Contacts:')
    for key, value in contact.items():
        print(f'Name: {key} ----- Number: {value}')
    print('\n')
    print('All Contacts are displayed')
    print('---------------------------------------------------------------------')
    
    
main()
```
