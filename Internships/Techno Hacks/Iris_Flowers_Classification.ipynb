# Iris Flowers Classification

### Import Libraries

```python
import warnings
warnings.simplefilter("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, confusion_matrix
```

### Import Data

```python
data_file_path = '/content/iris.data'

data = []

with open(data_file_path, 'r') as file:

    for line in file:
        columns = line.strip().split(',')

        data.append(columns)
```

```python
df = pd.DataFrame(data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

df.head()
```

```python
df.tail()
```

```python
df.info()
```

### Preprocessing

```python
last_row_index = df.index[-1]

df = df.drop(last_row_index)
```

```python
df['sepal_length'] = df['sepal_length'].astype(float, errors='raise')
df['sepal_width'] = df['sepal_width'].astype(float, errors='raise')
df['petal_length'] = df['petal_length'].astype(float, errors='raise')
df['petal_width'] = df['petal_width'].astype(float, errors='raise')
```

```python
df.info()
```

```python
df.describe()
```

```python
df['class'].value_counts()
```

### Extract input and output variables

```python
x = df.iloc[:,:4]

y = df.iloc[:,4]
```

```python
x
```

```python
y
```

### Split the Data Into Train and Test Datasets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Model Training and Testing

```python
Model = LogisticRegression()

Model.fit(x_train,y_train)

y_pred = Model.predict(x_test)
```

```python
y_pred
```

### Model Evaluation

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)*100

print("Accuracy of the model is {:.2f}".format(accuracy))
```
